```python
import matplotlib.pyplot as plt
import pandas as pd
animal='Animal.csv'
df=pd.read_csv(animal)
plt.figure(figsize=(5,5))
type_count=df['Habitat'].value_counts()
plt.pie(type_count,labels=type_count.index,autopct='%1.1f%%',startangle=90)
plt.show()
```

```python
plt.figure(figsize=(12,6))
hab_counts= df['Habitat'].value_counts()
hab_counts.plot(kind='bar',color='#34abeb')
plt.xlabel("Habitat")
plt.ylabel("Number of Animals")
plt.xticks(rotation=30,ha="right")
plt.grid(True)
plt.tight_layout()
plt.show()
```
